```python
bac = web.DataReader('BAC', 'yahoo', '2006-01-01', '2016-01-01')
c = web.DataReader('C', 'yahoo', '2006-01-01', '2016-01-01')
gs = web.DataReader('GS', 'yahoo', '2006-01-01', '2016-01-01')
jpm = web.DataReader('JPM', 'yahoo', '2006-01-01', '2016-01-01')
ms = web.DataReader('MS', 'yahoo', '2006-01-01', '2016-01-01')
wfc = web.DataReader('WFC', 'yahoo', '2006-01-01', '2016-01-01')
```

```python
bac.columns = pd.MultiIndex.from_product([['BAC'], bac.columns])
c.columns = pd.MultiIndex.from_product([['C'], c.columns])
gs.columns = pd.MultiIndex.from_product([['GS'], gs.columns])
jpm.columns = pd.MultiIndex.from_product([['JPM'], jpm.columns])
ms.columns = pd.MultiIndex.from_product([['MS'], ms.columns])
wfc.columns = pd.MultiIndex.from_product([['WFC'], wfc.columns])
```

** Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

** Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list. Also pay attention to what axis you concatenate on.**

```python
bank_stocks = pd.concat([bac, c, gs, jpm, ms, wfc], axis=1)
bank_stocks.head()
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

# EDA

Let's explore the data a bit! Before continuing, I encourage you to check out the documentation on [Multi-Level Indexing](http://pandas.pydata.org/pandas-docs/stable/advanced.html) and [Using .xs](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.xs.html).
Reference the solutions if you can not figure out how to use .xs(), since that will be a major part of this project.

** What is the max Close price for each bank's stock throughout the time period?**

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = bank_stocks.xs(key='Close',axis=1,level='Stock Info').pct_change()
returns.columns = ['BAC Return', 'C Return', 'GS Return', 'JPM Return', 'MS Return', 'WFC Return']
returns.head()
```

** Create a pairplot using seaborn of the returns dataframe

```python
sns.pairplot(returns[1:])
```

** Using this returns DataFrame, figure out on what dates each bank stock had the best and worst single day returns. You should notice that 4 of the banks share the same day for the worst drop, did anything significant happen that day?**

```python
#Best single day returns
bestReturn = []
for bank in returns.columns:
    bestReturn.append(returns[returns[bank] == returns[bank].max()].index[0])

pd.Series(index=returns.columns, data=bestReturn)
```

```python
#Worst single day returns
worstReturn = []
for bank in returns.columns:
    worstReturn.append(returns[returns[bank] == returns[bank].min()].index[0])

worstReturn
pd.Series(index=returns.columns, data=worstReturn)
```

** Take a look at the standard deviation of the returns, which stock would you classify as the riskiest over the entire time period? Which would you classify as the riskiest for the year 2015?**

```python
#Standard deviation of returns
returns.std()

#Citigroup the riskiest
```

```python
returns.loc['2015':'2015'].std()
#Morgan Stanley the riskiest
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
plt.figure(figsize=(12,6))
bank_stocks['BAC','Close'].loc['2007-12-01':'2008'].rolling(window=30).mean().plot(label='30 Day Avg')
bank_stocks['BAC','Close'].loc['2008':'2008'].plot(label='BAC CLOSE')
plt.legend()
```

** Create a heatmap of the correlation between the stocks Close Price.**

```python
sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(), annot=True, cmap = 'coolwarm')
```

** Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(), annot=True, cmap = 'coolwarm')
```
